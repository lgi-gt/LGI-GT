```python
import matplotlib.pyplot as plt 

plt.rcParams['xtick.bottom'] = plt.rcParams['xtick.labelbottom'] = False
plt.rcParams['xtick.top'] = plt.rcParams['xtick.labeltop'] = True 

nm = [
    [ 0.7830, 0.7752, 0.7210, 0.6014], 
    [ 0.7688, 0.7724, 0.7547, 0.6028], 
    [ 0.6850, 0.7140, 0.5798, 0.5009], 
    [ 0.3969, 0.4075, 0.3892, 0.3665], 
] 

plt.matshow(nm, cmap=plt.cm.Greens, alpha=0.8) 
plt.show() 
```

```python
import numpy as np 
```

```python
tag_y = ["1", "2", "4", "8"] 
tag_x = ["1", "2", "4", "8"] 
```

```python
fig, ax = plt.subplots() 

ax.set_xticks(np.arange(len(tag_x))) 
ax.set_yticks(np.arange(len(tag_y))) 
ax.set_xticklabels(tag_x) 
ax.set_yticklabels(tag_y) 
ax.xaxis.set_label_position('top') 
im = ax.imshow(nm, cmap='YlGn', origin='upper') 

plt.colorbar(im) 
for i in range(len(tag_y)): 
    for j in range(len(tag_x)): 
        ax.text(j, i, nm[i][j], ha='center', va='center', color='black') 

default_font = {'weight': 'bold', 'size': 14}
plt.xlabel("m", default_font) 
plt.ylabel("n", default_font) 

plt.savefig('nm_cluster.pdf', dpi=600) 
```
