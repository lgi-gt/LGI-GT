```python
import matplotlib.pyplot as plt 
import numpy as np 

ax = plt.axes()
ax.spines['top'].set_visible(False)  
ax.spines['right'].set_visible(False) 
ax.spines['bottom'].set_visible(True) 
ax.spines['left'].set_visible(True)

labels = ['ogbg-molpcba (AP)', 'ogbg-code2 (F1 score)'] 

Mean = [0.2603, 0.1692] 
Sum = [0.2192, 0.1443] 
Original_CLS = [0.2924, 0.1691] 
Skip_CLS = [0.3040, 0.1948] 

bar_width = 0.12 
index_Mean = np.arange(len(labels)) 
index_Sum = index_Mean + 0.15 
index_Original_CLS = index_Sum + 0.15 
index_Skip_CLS = index_Original_CLS + 0.15 

plt.bar(index_Mean, height=Mean, width=bar_width, color='#B2B2FF', edgecolor='blue', label='Mean') 
plt.bar(index_Sum, height=Sum, width=bar_width, color='#FFB2B2', edgecolor='red', label='Sum') 
plt.bar(index_Original_CLS, height=Original_CLS, width=bar_width, color='#ECD9C6', edgecolor='#734D26', label='Original [CLS]') 
plt.bar(index_Skip_CLS, height=Skip_CLS, width=bar_width, color='#B5E61D', edgecolor='g', label='Skip [CLS]') 

for a, b in zip(index_Mean, Mean): 
    plt.text(a, b+0.01, '%.4f'%b, ha='center', va='bottom', fontsize=8) 

for a, b in zip(index_Sum, Sum): 
    plt.text(a, b+0.01, '%.4f'%b, ha='center', va='bottom', fontsize=8) 

for a, b in zip(index_Original_CLS, Original_CLS): 
    plt.text(a, b+0.01, '%.4f'%b, ha='center', va='bottom', fontsize=8) 

for a, b in zip(index_Skip_CLS, Skip_CLS): 
    plt.text(a, b+0.01, '%.4f'%b, ha='center', va='bottom', fontsize=8) 

plt.rcParams.update({'font.size': 10}) 
plt.yticks(fontsize=10)  
plt.xticks(index_Sum + 0.075, labels, fontsize=10) 
default_font = {'size': 10} 

plt.ylabel("Test Performance", font=default_font) 

plt.legend() 

plt.savefig('cls.pdf', dpi=600) 
```
